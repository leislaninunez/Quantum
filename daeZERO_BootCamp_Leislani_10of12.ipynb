### Challenge 7: Math Library
**Import the math library. Store any number in a variable. Print its square root. Then print whether the number is bigger or smaller than pi using an if/else statement.**

**What this code does:**
This code imports Python's built-in math library and uses a value stored inside it: math.pi. The import keyword loads the library and makes everything inside it available using dot notation. An if/else then compares a variable to math.pi and prints a different result depending on the comparison. This combines three concepts: imports, variables, and conditionals.

---

```python



import math

my_number = 4

if my_number > math.pi:
    print(my_number,"is greater than pi",math.pi)
else:
    print(my_number,"is this less than pi",math.pi)

```

### Challenge 9: NumPy Arrays
**Import numpy as np. Create an array of at least 5 temperatures in Fahrenheit. Convert all of them to Celsius in one line using: celsius = (fahrenheit - 32) * 5/9. Print both arrays and the average temperature in Celsius.**

**What this code does:**
This code imports NumPy — a third-party library for numerical math — and uses it to convert an entire list of temperatures at once. np.array() creates a NumPy array, which unlike a regular Python list, supports bulk mathematical operations. Applying the conversion formula (fahrenheit - 32) * 5/9 to a NumPy array converts every value simultaneously — no loop required. np.mean() then calculates the average across all values in one call. This is why scientists and engineers worldwide rely on NumPy.

---

```python
# Convert Fahrenheit to Celsius using NumPy

import numpy as np

fahrenheit = np.array([23, 46, 52, 86, 73])
celsius=(fahrenheit-32)*5/9
print(celsius)
```
